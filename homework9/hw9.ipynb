# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
def plot_functions():
    x = np.linspace(1, 10, 10) #range of values for the domain, 10 spaces from 1-10
    f_x = x**2
    g_x = np.sqrt(x)  #input for f(x) and g(x)


    #plotting 
    plt.figure(figsize =(10,8))
    plt.plot(x, f_x, color = "green", linestyle = '--', label = "f(x) = x^2")
    plt.plot(x, g_x, color = "purple", linestyle = '--', label = "g(x) = sqrt(x)")

    #titles 
    plt.title("Plot of f(x) = x^2 and g(x) = sqrt(x)")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")

    #adding legend 
    plt.legend()
    #showing plot 
    plt.show()

    return f_x.tolist(), g_x.tolist() #returns lists of f(x) and g(x) 

f_values, g_values = plot_functions() #call function
               

```

```python
# Import your module and/or your code here
import numpy as np
import matplotlib.pyplot as plt

#generating data and setting bounds \
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x,y)
z = x**2 + y**3

#creating the checkered pattern for the plot
checkered = (np.floor(x) + np.floor(y)) % 2 #this alternates it by 1 and 0

#setting up my plot to be a 3d figure
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111, projection = "3d")

#plotting the checkered surface 
ax.plot_surface(x, y, z, facecolors=plt.cm.binary(checkered), rstride=1, cstride=1) 
#had to ask ChatGpt for this part: facecolors=plt.cm.binary(checkered) applies binary colormap to alternate colors based on values in checkered
#setting rstride=1 and cstride=1 displays grid without skipping rows and columns

#labeling
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Checkered Plot of z = x^2 + y^3')

plt.show()
```
